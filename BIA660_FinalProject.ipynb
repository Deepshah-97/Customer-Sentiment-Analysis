We will be scraping 2 popular products from Amazon
Product 1- Beats Studio Noise Cancelling Earbuds
Product 2- Apple AirPods Pro 2nd generation

Amazon has a limit on 10 pages of reviews for each product. To circumvernt this we shall take 10 pages, of each products number of stars

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import pandas as pd

def get_amazon_reviews(base_url, start_page, end_page):
    driver = webdriver.Chrome()
    review_list=[]
    title_list=[]
    rating_list=[]

    for page in range(start_page, end_page + 1):
        url = f"{base_url}&pageNumber={page}"
        driver.get(url)
        time.sleep(20)  # Allow some time for the page to load

        soup = BeautifulSoup(driver.page_source, 'html.parser')
        reviews = soup.find_all('div', {'data-hook': 'review'})
        print(len(reviews))

        try:
            for review in reviews:
                username = review.find('span', {'class': 'a-profile-name'}).text.strip()
                text = review.find('span', {'data-hook': 'review-body'}).text.strip()
                rating = review.find('span', {'class': 'a-icon-alt'}).text.strip()

                review_list.append({'Username': username, 'Review': text, 'Rating': rating})
                print(f"Username: {username}, Rating: {rating}")
                print(f"Review: {text}\n")
                

        except Exception as e:
            print(f"Error: {e}")

    driver.quit()
    return review_list




```

```python
#urls of 2 products
base_url_1 = 'https://www.amazon.com/Beats-Studio-Cancelling-Earbuds-Built-Bluetooth-Headphones/product-reviews/B096SV8SJG/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
base_url_2 = 'https://www.amazon.com/Apple-Generation-Cancelling-Transparency-Personalized/product-reviews/B0CHWRXH8B/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'


```
